```python
# Standard imports - do not modify
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)
```

## Part 1: Load and Explore Data

We'll work with a simple synthetic dataset.

```python
# Generate synthetic data
n = 100
X = np.random.uniform(0, 10, n)
y = 2.5 * X + 5 + np.random.normal(0, 2, n)  # y = 2.5x + 5 + noise

# Create DataFrame
df = pd.DataFrame({'X': X, 'y': y})
df.head()
```

### Task 1.1: Basic Statistics (5 points)

Calculate the mean of `X` and the mean of `y`. Store them in variables `mean_x` and `mean_y`.

```python
# YOUR CODE HERE
mean_x = None
mean_y = None

mean_x = df['X'].mean()
mean_y = df['y'].mean()

print(f"Mean of X: {mean_x}")
print(f"Mean of y: {mean_y}")
```

### Task 1.2: Scatter Plot (5 points)

Create a scatter plot of X vs y. Label your axes.

```python
# YOUR CODE HERE

plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("A scatter plot of X vs y")
plt.tight_layout()
plt.show()
```

## Part 2: Simple Linear Regression

### Task 2.1: Calculate Slope and Intercept (10 points)

Using the formulas from class, calculate the slope (`beta_1`) and intercept (`beta_0`) for simple linear regression.

Recall:
- $\hat{\beta}_1 = \frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n}(x_i - \bar{x})^2}$
- $\hat{\beta}_0 = \bar{y} - \hat{\beta}_1 \bar{x}$

```python
# YOUR CODE HERE
beta_1 = None  # slope
beta_0 = None  # intercept

beta_1 = np.sum(((X - mean_x) * (y - mean_y))) / np.sum(((X - mean_x) ** 2))
beta_0 = mean_y - beta_1 * mean_x

print(f"Slope (beta_1): {beta_1}")
print(f"Intercept (beta_0): {beta_0}")
```

### Task 2.2: Make Predictions (5 points)

Using your calculated `beta_0` and `beta_1`, create predictions `y_pred` for all values in `X`.

```python
# YOUR CODE HERE
y_pred = None

y_pred = beta_1 * X + beta_0

print(f"First 5 predictions: {y_pred[:5]}")
```

### Task 2.3: Plot Regression Line (5 points)

Create a scatter plot of the data with your regression line overlaid.

```python
# YOUR CODE HERE
plt.scatter(X, y, label = "data")
ind = np.argsort(X)
plt.plot(X[ind], y_pred[ind], label = "regression line")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.tight_layout()
plt.show()
```

## Part 3: Model Evaluation

### Task 3.1: Calculate MSE (5 points)

Calculate the Mean Squared Error of your predictions.

$MSE = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2$

```python
# YOUR CODE HERE
mse = None

mse = np.sum((y - y_pred) ** 2) / len(y_pred)

print(f"Mean Squared Error: {mse}")
```

### Task 3.2: Calculate R-squared (5 points)

Calculate the R-squared value.

$R^2 = 1 - \frac{SS_{res}}{SS_{tot}} = 1 - \frac{\sum(y_i - \hat{y}_i)^2}{\sum(y_i - \bar{y})^2}$

```python
# YOUR CODE HERE
r_squared = None

r_squared = 1 - np.sum((y - y_pred) ** 2) / np.sum((y - mean_y) ** 2)

print(f"R-squared: {r_squared}")
```
